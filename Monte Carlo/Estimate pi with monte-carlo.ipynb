```python
import numpy as np
import random
import matplotlib.pyplot as plt

class PiEstimation():
    def __init__(self,plot=False):
        self.plot=plot

    def classify_point(self,x,y):
        return True if (pow(x,2)+ pow(y,2))<=1 else False 
    
    def estimate_pi(self,*trials):
        if self.plot is True:
            colors=['red','blue']
            col=2
            rows=int(np.ceil(len(trials[0])/col))
            fig, axs = plt.subplots(rows,col, figsize=(15, 15*(rows/col)), facecolor='w', edgecolor='k')
            fig.subplots_adjust(hspace = .5, wspace=.15)
            axs = axs.ravel()
        for idx,n_trials in enumerate(trials[0]):
            circle_points=0
            real_pi=3.14159265
            points=[]
            results=[]
            for i in range (n_trials):
                x=random.uniform(-1,1)
                y=random.uniform(-1,1)
                if self.classify_point(x,y) is True:
                    circle_points+=1
                    r= 1
                else:
                    r= 0
                cords=(x,y)
                points.append(cords)
                results.append(r)
                if self.plot is True:
                    axs[idx].scatter(x, y, color=colors[r])                                                   
            '''pi formula:
            area of circle= pi*pow(r,2)
            aera of a square=pow(2*r,2)==> 4*pow(r,2)
            area of circle/aera of a square = pi*pow(r,2)/4*pow(r,2) ==>pi/4
            pi=4*(area of circle/aera of a square)
            '''
            pi=round(4*(circle_points/n_trials),4)
            error=np.round(abs(pi-real_pi),4)
            axs[idx].set_title("# of trials: {},\nEstimated Pi is {}, Error is: {}".format(n_trials,pi,error),size=12, family='serif',position=(0.5,1.1))
        plt.show()
        return pi,error

# ===========================  create number of trials list:  ===========================
# interval=10
# trials=[]
# for i in range (0,6):
#     interval+=10
#     trials.append(int(interval))

trials=[10,100,1000,3000,10000,25000]

p=PiEstimation(plot=True)
pi,error=p.estimate_pi(trials)
# plt.plot(trials,errors)
# plt.show()
```
